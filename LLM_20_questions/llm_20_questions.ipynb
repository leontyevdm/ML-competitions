```python
%%writefile /tmp/submission/main.py
import re
import os
KAGGLE_AGENT_PATH = "/kaggle_simulations/agent/"
VERBOSE = False
if not os.path.exists(KAGGLE_AGENT_PATH + "weights"):
    KAGGLE_AGENT_PATH = "/tmp/submission/"
    VERBOSE = True

#################
# LOAD MODEL INTO MEMORY
import sys, torch
from transformers import AutoTokenizer, AutoModelForCausalLM
sys.path.insert(0, f"{KAGGLE_AGENT_PATH}lib")
model_not_loaded=True
if model_not_loaded:
    model = AutoModelForCausalLM.from_pretrained(
        f"{KAGGLE_AGENT_PATH}weights/",
        torch_dtype = torch.float16,
        device_map = "auto",
        trust_remote_code = True,
    )
    model_not_loaded = False
    tokenizer = AutoTokenizer.from_pretrained(f"{KAGGLE_AGENT_PATH}weights/")

import pandas as pd, numpy as np
saved_word = None
info = ''
ALPHA_OPPONENT = ''
guesses = []
firstword = ''
dumb_counter = 0
pop_words = False
q_list_counter = 0
questions_list = ['Is the keyword something that is typically found indoors?',
                 'Is the keyword man-made?']
def parse_alpha_question(question):
        # Regex to respond to alpha questions, includes all the 4 variants from the previous version:
        match = re.search(r'keyword.*(?:come before|precede) \"([^\"]+)\" .+ order\?$', question)
        if match:
            # Alpha question matched
            testword = match.group(1)
            return testword
        else:
            return None
        
def ask():
    global info
    prompt = f"We are playing 20 questions. You need to guess the keyword. Here is what you know about keyword: {info}. Generate next simple broad question to guess the keyword."
    messages = [
        {"role": "system", "content": "The question must start with 'Is the keyword' or 'Does the keyword' or 'May the keyword'. Don't ask questions about information that is already provided. Generate the question and nothing else."},
        {"role": "user", "content": prompt}
    ]
    #"Generate the only question starting with 'Is the keyword' or 'Does the keyword' to guess the keyword and nothing else."
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )
    model_inputs = tokenizer([text], return_tensors="pt").to('cuda')
    pad_token_id = tokenizer.pad_token_id
    if pad_token_id is None:
        pad_token_id = tokenizer.eos_token_id
    generated_ids = model.generate(
        model_inputs.input_ids,
        attention_mask = model_inputs.attention_mask,
        pad_token_id = pad_token_id,
        max_new_tokens=100,
    )
    generated_ids = [
        output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
    ]
    response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
    #response = response[response.find('Is the keyword'):]
    #print(response)
    try:
        response = response[:response.find('?') + 1]
    except:
        pass
    del generated_ids
    if response == "":
        response = 'Does the keyword consist of more than 7 letters?'
        
    return response
def guess_default(answers):
    
    global info
    global guesses
    #for i in range(len(answers) - 1):
    #    info += 'It is not ' + guesses[i] + '. '
    prompt = f"We are playing 20 questions. You need to guess the keyword. Here is what you know about keyword: {info} Try to guess the keyword."
    messages = [
        {"role": "system", "content": "Based on provided information, guess the keyword. The keyword may consist of two or more words. Write the keyword only and nothing else."},
        {"role": "user", "content": prompt}
    ]
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )
    model_inputs = tokenizer([text], return_tensors="pt").to('cuda')
    
    pad_token_id = tokenizer.pad_token_id
    if pad_token_id is None:
        pad_token_id = tokenizer.eos_token_id
        
    generated_ids = model.generate(
        model_inputs.input_ids,
        attention_mask = model_inputs.attention_mask,
        pad_token_id = pad_token_id,
        max_new_tokens=10,
    )

    generated_ids = [
        output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
    ]

    response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
    if response.endswith('assistant'):
        response = response[:-len('assistant')]
    theme = response.lower()
    response = response.lower()
    if response in guesses:
        response = guess(answers)
    if response == None:
        response = guess_multi(answers, theme)
    return response

def guess_multi(answers, theme):
    
    global info
    global guesses
    for i in range(len(answers) - 1):
        info += 'It is not ' + guesses[i] + '. '
    prompt = f"We are playing 20 questions. You need to guess the keyword. Here is what you know about keyword: {info}. The keyword may consist of two or more words like 'car engine' or 'electric razor'. Try to guess the keyword."
    messages = [
        {"role": "system", "content": "Based on provided information, guess the keyword. The keyword may consist of two or more words. Write the keyword only and nothing else."},
        {"role": "user", "content": prompt}
    ]
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )
    model_inputs = tokenizer([text], return_tensors="pt").to('cuda')
    
    pad_token_id = tokenizer.pad_token_id
    if pad_token_id is None:
        pad_token_id = tokenizer.eos_token_id
        
    generated_ids = model.generate(
        model_inputs.input_ids,
        attention_mask = model_inputs.attention_mask,
        pad_token_id = pad_token_id,
        max_new_tokens=5,
    )

    generated_ids = [
        output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
    ]

    response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
    if response.endswith('assistant'):
        response = response[:-len('assistant')]
    if VERBOSE:
        print(f"### {prompt}")
        print(response)
    response = response.lower()
    return response


def guess(answers):
    global info
    global guesses
    
    prompt = f"We are playing 20 questions. You need to guess the keyword. Here is what you know about keyword: {info} Try to guess the keyword."
    messages = [
        {"role": "system", "content": "Based on provided information, guess the keyword. The keyword may consist of two or more words. Write the keyword only and nothing else."},
        {"role": "user", "content": prompt}
    ]
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )
    model_inputs = tokenizer([text], return_tensors="pt").to('cuda')
    
    pad_token_id = tokenizer.pad_token_id
    if pad_token_id is None:
        pad_token_id = tokenizer.eos_token_id
        
    gen_ids = model.generate(
        model_inputs.input_ids,
        attention_mask=model_inputs.attention_mask,
        pad_token_id=pad_token_id,
        max_new_tokens=10,
        #do_sample=False,  # Set to False for greedy decoding
        #num_beams=20,     # Set the number of beams to 20 to get top variants
        #num_return_sequences=20  # Th
        do_sample=True,
        top_k=30,
        top_p=0.99,
        num_return_sequences=10,
        temperature=1.0,
    )
    generated_ids = [gen_ids[i][len(model_inputs.input_ids):] for i in range(10)]
    del model_inputs
    torch.cuda.empty_cache()
    response = [tokenizer.decode(generated_ids[i], skip_special_tokens=True) for i in range(10)]
    for i in range(len(response)):
        response[i] = response[i][response[i].find('assistant\n\n') + 11:].lower()
        response[i] = re.sub(r'[^a-z\s]', '', response[i])
    resp_dict = dict()
    for i in range(len(response)):
        if response[i] in resp_dict:
            resp_dict[response[i]] += 1
        else:
            resp_dict[response[i]] = 1
    resp = list(dict(sorted(resp_dict.items(), key=lambda item: item[1], reverse=True)).keys())
    print(resp)
    if VERBOSE:
        print(f"### {prompt}")
        print(response)
    del generated_ids, gen_ids
    del response
    torch.cuda.empty_cache()
    response = None
    for r in resp:
        if r.lower() not in guesses:
            response = r
            break
    return response

def get_yes_no(q,keyword):
    global keywords, VERBOSE
    prompt = f"We are playing 20 questions. You will answer questions about the keyword. The keyword is '{keyword}'. The keyword has first letter '{keyword[0]}'. {q}"
    messages = [
        {"role": "system", "content": "Answer yes or no to the following question and nothing else."},
        {"role": "user", "content": prompt}
    ]
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )
    model_inputs = tokenizer([text], return_tensors="pt").to('cuda')
    
    pad_token_id = tokenizer.pad_token_id
    if pad_token_id is None:
        pad_token_id = tokenizer.eos_token_id
        
    generated_ids = model.generate(
        model_inputs.input_ids,
        attention_mask = model_inputs.attention_mask,
        pad_token_id = pad_token_id,
        max_new_tokens=1
    )
    generated_ids = [
        output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
    ]

    response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
    if not "yes" in response.lower(): response = "no"
    else: response = "yes"
        
    if VERBOSE:
        print(f"### {prompt}")
        
    return response


firstword = 'cooling'
def search_pair():
    global firstword
    prompt = f"Your task is to suggest suitable noun to fill in the [MASK] token to get a name of a thing or an object. For example, for query 'bird [MASK]' possible answers may be 'cage' or 'seed'. Fill in the gap token here:'{firstword} [MASK]'"
    messages = [
            {"role": "system", "content": "The output must be only one word and nothing else"},
            {"role": "user", "content": prompt}
    ]
    #"Generate the only question starting with 'Is the keyword' or 'Does the keyword' to guess the keyword and nothing else."
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )
    model_inputs = tokenizer([text], return_tensors="pt").to('cuda')
    pad_token_id = tokenizer.pad_token_id
    if pad_token_id is None:
        pad_token_id = tokenizer.eos_token_id
    gen_ids = model.generate(
        model_inputs.input_ids,
        attention_mask=model_inputs.attention_mask,
        pad_token_id=pad_token_id,
        max_new_tokens=5,
        #do_sample=False,  # Set to False for greedy decoding
        #num_beams=20,     # Set the number of beams to 20 to get top variants
        #num_return_sequences=20  # Th
        do_sample=True,
        top_k=16,
        top_p=0.99999,
        num_return_sequences=50,
        temperature=0.95,
    )
    generated_ids = [gen_ids[i][len(model_inputs.input_ids):] for i in range(50)]
    response = [tokenizer.decode(generated_ids[i], skip_special_tokens=True) for i in range(50)]
    for i in range(len(response)):
        response[i] = response[i][response[i].find('assistant\n\n') + 11:].lower()
        response[i] = re.sub(r'[^a-z\s]', '', response[i])
    resp_dict = dict()
    for i in range(len(response)):
        if response[i] in resp_dict:
            resp_dict[response[i]] += 1
        else:
            resp_dict[response[i]] = 1
    resp = list(dict(sorted(resp_dict.items(), key=lambda item: item[1], reverse=True)).keys())[:12]
    del generated_ids
    print(resp)
    return resp
############
# MAIN AGENT FUNCTION
def agent_beta(obs, cfg):
    global info, guesses, ALPHA_OPPONENT, q_list_counter, questions_list
    global is_place, is_city, is_country
    # GENERATE RESPONSE
    if obs.turnType == "ask":
        info = 'The keyword is a thing that exists in real world.'
        answers_length = len(obs.answers)
        for i in range(1, answers_length):
            if obs.questions[i].startswith('Is the keyword'):
                if obs.answers[i] == 'yes':
                    info += 'The keyword is' + obs.questions[i][14:-1]  + '. '
                else:
                    info += 'The keyword is not' + obs.questions[i][14:-1] + '. '
            elif obs.questions[i].startswith('Does the keyword'):
                if obs.answers[i] == 'yes':
                    info += 'The keyword' + obs.questions[i][16:-1]  + '. '
                else:
                    info += 'The keyword does not' + obs.questions[i][16:-1]  + '. '
            elif obs.questions[i].startswith('May the keyword'):
                if obs.answers[i] == 'yes':
                    info += 'The keyword may' + obs.questions[i][16:-1]  + '. '
                else:
                    info += 'The keyword may not' + obs.questions[i][16:-1]  + '. '
            else:
                info += ' Answer for question "' + obs.questions[i] + '" is ' +  obs.answers[i]  + '. '
        for i in range(len(guesses)):
            info += f"It is not '{guesses[i]}'. "
            
        if q_list_counter < len(questions_list):
            response = questions_list[q_list_counter]
            q_list_counter += 1
        else:
            response = ask()
    elif obs.turnType == "guess":
        info = 'The keyword is a thing. '
        answers_length = len(obs.answers)
        for i in range(1, answers_length):
            if obs.questions[i].startswith('Is the keyword'):
                if obs.answers[i] == 'yes':
                    info += 'The keyword is' + obs.questions[i][14:-1]  + '. '
                else:
                    info += 'The keyword is not' + obs.questions[i][14:-1] + '. '
            elif obs.questions[i].startswith('Does the keyword'):
                if obs.answers[i] == 'yes':
                    info += 'The keyword' + obs.questions[i][16:-1]  + '. '
                else:
                    info += 'The keyword does not' + obs.questions[i][16:-1]  + '. '
            elif obs.questions[i].startswith('May the keyword'):
                if obs.answers[i] == 'yes':
                    info += 'The keyword may' + obs.questions[i][16:-1]  + '. '
                else:
                    info += 'The keyword may not' + obs.questions[i][16:-1]  + '. '
            else:
                info += ' Answer for question "' + obs.questions[i] + '" is ' +  obs.answers[i]  + '. '
     
        
        response = guess_default(obs.answers)
        guesses.append(response)
                
    else: #obs.turnType == "answer"
        HANDSHAKE = 'Is it Agent Alpha?'
        if obs.keyword.lower() in obs.questions[-1].lower():
            response = "yes"
        else:
            if HANDSHAKE == obs.questions[-1]:
                response = 'yes'
            else:
                testword = parse_alpha_question(obs.questions[-1])
                if testword is not None:
                    response = 'yes' if obs.keyword.lower() < testword else 'no'
                else:
                    response = get_yes_no(obs.questions[-1], obs.keyword)
            
    # DISPLAY ROLE
    if VERBOSE: 
        if obs.turnType == "answer": 
            print(f"Team 2 - Answerer - ### Agent LLAMA 8B ###")
        else:
            print(f"\nTeam 2 - Questioner - ### Agent LLAMA 8B ###")
        print(f"OUTPUT = '{response}'")

    return response


import random
import re

VERSION = 9
import pickle
with open(KAGGLE_AGENT_PATH + 'naive_216.pkl', 'rb') as f:
    wordlist = pickle.load(f)
for i in range(len(wordlist)):
    wordlist[i] = wordlist[i].lower()
# The list of all expected keywords in the game. It has to be updated before submitting your agent for the evaluation
# phase. Apply your own judgement. The longer the list the better coverage, but slower convergence. The coverage is
# arguably more important than convergence, so having many extra keywords is better than having missing words.
allwords = wordlist

last_alpha_yes = ''
last_alpha_no = ''


def answered_yes(obs, question):
    """ Check whether the question has been asked and the answer was 'yes'. """
    try:
        ix = obs.questions.index(question)
        return obs.answers[ix] == 'yes'
    except (ValueError, IndexError):
        return False


class AgentAlpha:
    global allwords
    # Keep it fixed, for compatibility
    HANDSHAKE = 'Is it Agent Alpha?'

    # This is our search space, it gets modified, so it needs a
    # reinit when testing with a series of games in a row
    keywords = sorted(allwords)

    @staticmethod
    def parse_alpha_question(question):
        # Regex to respond to alpha questions, includes all the 4 variants from the previous version:
        match = re.search(r'keyword.*(?:come before|precede) \"([^\"]+)\" .+ order\?$', question)
        if match:
            # Alpha question matched
            testword = match.group(1)
            return testword
        else:
            return None

    @staticmethod
    def play(obs, cfg):
        global last_alpha_yes, last_alpha_no, guesses, firstword, ALPHA_OPPONENT, pop_words
        if obs.turnType == 'ask':

            if len(obs.questions) == 0:
                return AgentAlpha.HANDSHAKE

            elif answered_yes(obs, AgentAlpha.HANDSHAKE) and AgentAlpha.keywords:
                ALPHA_OPPONENT = 'yes'
                testword = AgentAlpha.keywords[len(AgentAlpha.keywords) // 2]
                response = f'Does the keyword (in lowercase) precede "{testword}" in alphabetical order?'

            elif not AgentAlpha.keywords:
                response = 'No more alpha'
            else:
                response = None

        elif obs.turnType == 'answer':
            if AgentAlpha.HANDSHAKE == obs.questions[-1]:
                response = 'yes'
                #response = 'no' #declined

            else:
                testword = AgentAlpha.parse_alpha_question(obs.questions[-1])
                if testword is not None:
                    response = 'yes' if obs.keyword.lower() < testword else 'no'
                else:
                    # Not an alpha question, proceed with another approach
                    response = None

        elif obs.turnType == 'guess':

            # If the last question was a handshake, give a special guess
            if obs.questions[-1] == AgentAlpha.HANDSHAKE:
                # That's the handshake round...
                if obs.answers[-1] == 'yes':
                    ALPHA_OPPONENT = 'yes'
                    response = "plastic windows"
                else:
                    ALPHA_OPPONENT = 'no'
                    response = f'eye'

            else:
                # Bisect the space based on the last question and answer
                testword = AgentAlpha.parse_alpha_question(obs.questions[-1])
                if testword:
                    if obs.answers[-1] == 'yes':
                        last_alpha_yes = testword
                        AgentAlpha.keywords = [k for k in AgentAlpha.keywords if k < testword]
                    else:
                        last_alpha_no = testword
                        AgentAlpha.keywords = [k for k in AgentAlpha.keywords if not k < testword]
                firstword = last_alpha_no
                print('borders ', last_alpha_no, last_alpha_yes)
                if AgentAlpha.keywords:
                    # We pop from the front, but it doesn't matter much from which side we pop
                    if len(AgentAlpha.keywords) == 1 and (last_alpha_no == AgentAlpha.keywords[0] or last_alpha_yes == AgentAlpha.keywords[0]):
                        response = AgentAlpha.keywords.pop(0)
                    elif pop_words == False:
                        response = AgentAlpha.keywords[0]
                    else:
                        response = AgentAlpha.keywords.pop(0)
                else:
                    # Keywords exhausted, use something else
                    response = None
            guesses.append(response)
        else:
            assert False, f'Unexpected turnType: {obs.turnType}'

        return response
variants = []
def agent_dumb(obs, cfg):
    global variants
    # GENERATE RESPONSE
    if obs.turnType == "ask":
        response = f'Is it {variants[0]}?'
    elif obs.turnType == "guess":
        response = variants.pop(0)
    else: #obs.turnType == "answer"
        response = 'yes'
    return response

def agent_fn(obs, cfg):
    global variants, pop_words, guesses
    global ALPHA_OPPONENT, firstword, allwords, saved_word
    response = None
    if ALPHA_OPPONENT == '' or ALPHA_OPPONENT == 'yes':
        response = AgentAlpha.play(obs, cfg)
    if (response == 'No more alpha') or (ALPHA_OPPONENT == 'yes' and response is None) and not pop_words:
        #ALPHA_OPPONENT = 'dumb'
        print('NO MORE ALPHA')
        saved_word = firstword
        variants = search_pair()
        variants = [firstword + ' ' + variants[i].lower() for i in range(len(variants))]
        AgentAlpha.keywords = sorted(variants)
        pop_words = True
        response = AgentAlpha.play(obs, cfg)
    if pop_words and response is None:
        response = 'no'
    if response is None:
        response = agent_beta(obs, cfg)
    return response
```

```python
import re
import os
KAGGLE_AGENT_PATH = "/kaggle_simulations/agent/"
VERBOSE = False
if not os.path.exists(KAGGLE_AGENT_PATH + "weights"):
    KAGGLE_AGENT_PATH = "/tmp/submission/"
    VERBOSE = True

#################
# LOAD MODEL INTO MEMORY
import sys, torch
from transformers import AutoTokenizer, AutoModelForCausalLM
sys.path.insert(0, f"{KAGGLE_AGENT_PATH}lib")
if model_not_loaded:
    model = AutoModelForCausalLM.from_pretrained(
        f"{KAGGLE_AGENT_PATH}weights/",
        torch_dtype = torch.float16,
        device_map = "auto",
        trust_remote_code = True,
    )
    model_not_loaded = False
    tokenizer = AutoTokenizer.from_pretrained(f"{KAGGLE_AGENT_PATH}weights/")

import pandas as pd, numpy as np
saved_word = None
info = ''
ALPHA_OPPONENT = ''
guesses = []
firstword = ''
dumb_counter = 0
pop_words = False
q_list_counter = 0
questions_list = ['Is the keyword something that is typically found indoors?',
                 'Is the keyword man-made?']
def parse_alpha_question(question):
        # Regex to respond to alpha questions, includes all the 4 variants from the previous version:
        match = re.search(r'keyword.*(?:come before|precede) \"([^\"]+)\" .+ order\?$', question)
        if match:
            # Alpha question matched
            testword = match.group(1)
            return testword
        else:
            return None
        
def ask():
    global info
    prompt = f"We are playing 20 questions. You need to guess the keyword. Here is what you know about keyword: {info}. Generate next simple broad question to guess the keyword."
    messages = [
        {"role": "system", "content": "The question must start with 'Is the keyword' or 'Does the keyword' or 'May the keyword'. Don't ask questions about information that is already provided. Generate the question and nothing else."},
        {"role": "user", "content": prompt}
    ]
    #"Generate the only question starting with 'Is the keyword' or 'Does the keyword' to guess the keyword and nothing else."
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )
    model_inputs = tokenizer([text], return_tensors="pt").to('cuda')
    pad_token_id = tokenizer.pad_token_id
    if pad_token_id is None:
        pad_token_id = tokenizer.eos_token_id
    generated_ids = model.generate(
        model_inputs.input_ids,
        attention_mask = model_inputs.attention_mask,
        pad_token_id = pad_token_id,
        max_new_tokens=100,
    )
    generated_ids = [
        output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
    ]
    response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
    #response = response[response.find('Is the keyword'):]
    #print(response)
    try:
        response = response[:response.find('?') + 1]
    except:
        pass
    del generated_ids
    if response == "":
        response = 'Does the keyword consist of more than 7 letters?'
        
    return response
def guess_default(answers):
    
    global info
    global guesses
    #for i in range(len(answers) - 1):
    #    info += 'It is not ' + guesses[i] + '. '
    prompt = f"We are playing 20 questions. You need to guess the keyword. Here is what you know about keyword: {info} Try to guess the keyword."
    messages = [
        {"role": "system", "content": "Based on provided information, guess the keyword. The keyword may consist of two or more words. Write the keyword only and nothing else."},
        {"role": "user", "content": prompt}
    ]
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )
    model_inputs = tokenizer([text], return_tensors="pt").to('cuda')
    
    pad_token_id = tokenizer.pad_token_id
    if pad_token_id is None:
        pad_token_id = tokenizer.eos_token_id
        
    generated_ids = model.generate(
        model_inputs.input_ids,
        attention_mask = model_inputs.attention_mask,
        pad_token_id = pad_token_id,
        max_new_tokens=10,
    )

    generated_ids = [
        output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
    ]

    response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
    if response.endswith('assistant'):
        response = response[:-len('assistant')]
    theme = response.lower()
    response = response.lower()
    if response in guesses:
        response = guess(answers)
    if response == None:
        response = guess_multi(answers, theme)
    return response

def guess_multi(answers, theme):
    
    global info
    global guesses
    for i in range(len(answers) - 1):
        info += 'It is not ' + guesses[i] + '. '
    prompt = f"We are playing 20 questions. You need to guess the keyword. Here is what you know about keyword: {info}. The keyword may consist of two or more words like 'car engine' or 'electric razor'. Try to guess the keyword."
    messages = [
        {"role": "system", "content": "Based on provided information, guess the keyword. The keyword may consist of two or more words. Write the keyword only and nothing else."},
        {"role": "user", "content": prompt}
    ]
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )
    model_inputs = tokenizer([text], return_tensors="pt").to('cuda')
    
    pad_token_id = tokenizer.pad_token_id
    if pad_token_id is None:
        pad_token_id = tokenizer.eos_token_id
        
    generated_ids = model.generate(
        model_inputs.input_ids,
        attention_mask = model_inputs.attention_mask,
        pad_token_id = pad_token_id,
        max_new_tokens=5,
    )

    generated_ids = [
        output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
    ]

    response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
    if response.endswith('assistant'):
        response = response[:-len('assistant')]
    if VERBOSE:
        print(f"### {prompt}")
        print(response)
    response = response.lower()
    return response


def guess(answers):
    global info
    global guesses
    
    prompt = f"We are playing 20 questions. You need to guess the keyword. Here is what you know about keyword: {info} Try to guess the keyword."
    messages = [
        {"role": "system", "content": "Based on provided information, guess the keyword. The keyword may consist of two or more words. Write the keyword only and nothing else."},
        {"role": "user", "content": prompt}
    ]
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )
    model_inputs = tokenizer([text], return_tensors="pt").to('cuda')
    
    pad_token_id = tokenizer.pad_token_id
    if pad_token_id is None:
        pad_token_id = tokenizer.eos_token_id
        
    gen_ids = model.generate(
        model_inputs.input_ids,
        attention_mask=model_inputs.attention_mask,
        pad_token_id=pad_token_id,
        max_new_tokens=10,
        #do_sample=False,  # Set to False for greedy decoding
        #num_beams=20,     # Set the number of beams to 20 to get top variants
        #num_return_sequences=20  # Th
        do_sample=True,
        top_k=30,
        top_p=0.99,
        num_return_sequences=10,
        temperature=1.0,
    )
    generated_ids = [gen_ids[i][len(model_inputs.input_ids):] for i in range(10)]
    del model_inputs
    torch.cuda.empty_cache()
    response = [tokenizer.decode(generated_ids[i], skip_special_tokens=True) for i in range(10)]
    for i in range(len(response)):
        response[i] = response[i][response[i].find('assistant\n\n') + 11:].lower()
        response[i] = re.sub(r'[^a-z\s]', '', response[i])
    resp_dict = dict()
    for i in range(len(response)):
        if response[i] in resp_dict:
            resp_dict[response[i]] += 1
        else:
            resp_dict[response[i]] = 1
    resp = list(dict(sorted(resp_dict.items(), key=lambda item: item[1], reverse=True)).keys())
    print(resp)
    if VERBOSE:
        print(f"### {prompt}")
        print(response)
    del generated_ids, gen_ids
    del response
    torch.cuda.empty_cache()
    response = None
    for r in resp:
        if r.lower() not in guesses:
            response = r
            break
    return response

def get_yes_no(q,keyword):
    global keywords, VERBOSE
    prompt = f"We are playing 20 questions. You will answer questions about the keyword. The keyword is '{keyword}'. The keyword has first letter '{keyword[0]}'. {q}"
    messages = [
        {"role": "system", "content": "Answer yes or no to the following question and nothing else."},
        {"role": "user", "content": prompt}
    ]
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )
    model_inputs = tokenizer([text], return_tensors="pt").to('cuda')
    
    pad_token_id = tokenizer.pad_token_id
    if pad_token_id is None:
        pad_token_id = tokenizer.eos_token_id
        
    generated_ids = model.generate(
        model_inputs.input_ids,
        attention_mask = model_inputs.attention_mask,
        pad_token_id = pad_token_id,
        max_new_tokens=1
    )
    generated_ids = [
        output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
    ]

    response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
    if not "yes" in response.lower(): response = "no"
    else: response = "yes"
        
    if VERBOSE:
        print(f"### {prompt}")
        
    return response


firstword = 'cooling'
def search_pair():
    global firstword
    prompt = f"Your task is to suggest suitable noun to fill in the [MASK] token to get a name of a thing or an object. For example, for query 'bird [MASK]' possible answers may be 'cage' or 'seed'. Fill in the gap token here:'{firstword} [MASK]'"
    messages = [
            {"role": "system", "content": "The output must be only one word and nothing else"},
            {"role": "user", "content": prompt}
    ]
    #"Generate the only question starting with 'Is the keyword' or 'Does the keyword' to guess the keyword and nothing else."
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )
    model_inputs = tokenizer([text], return_tensors="pt").to('cuda')
    pad_token_id = tokenizer.pad_token_id
    if pad_token_id is None:
        pad_token_id = tokenizer.eos_token_id
    gen_ids = model.generate(
        model_inputs.input_ids,
        attention_mask=model_inputs.attention_mask,
        pad_token_id=pad_token_id,
        max_new_tokens=5,
        #do_sample=False,  # Set to False for greedy decoding
        #num_beams=20,     # Set the number of beams to 20 to get top variants
        #num_return_sequences=20  # Th
        do_sample=True,
        top_k=16,
        top_p=0.99999,
        num_return_sequences=50,
        temperature=0.95,
    )
    generated_ids = [gen_ids[i][len(model_inputs.input_ids):] for i in range(50)]
    response = [tokenizer.decode(generated_ids[i], skip_special_tokens=True) for i in range(50)]
    for i in range(len(response)):
        response[i] = response[i][response[i].find('assistant\n\n') + 11:].lower()
        response[i] = re.sub(r'[^a-z\s]', '', response[i])
    resp_dict = dict()
    for i in range(len(response)):
        if response[i] in resp_dict:
            resp_dict[response[i]] += 1
        else:
            resp_dict[response[i]] = 1
    resp = list(dict(sorted(resp_dict.items(), key=lambda item: item[1], reverse=True)).keys())[:12]
    del generated_ids
    print(resp)
    return resp
############
# MAIN AGENT FUNCTION
def agent_beta(obs, cfg):
    global info, guesses, ALPHA_OPPONENT, q_list_counter, questions_list
    global is_place, is_city, is_country
    # GENERATE RESPONSE
    if obs.turnType == "ask":
        info = 'The keyword is a thing that exists in real world.'
        answers_length = len(obs.answers)
        for i in range(1, answers_length):
            if obs.questions[i].startswith('Is the keyword'):
                if obs.answers[i] == 'yes':
                    info += 'The keyword is' + obs.questions[i][14:-1]  + '. '
                else:
                    info += 'The keyword is not' + obs.questions[i][14:-1] + '. '
            elif obs.questions[i].startswith('Does the keyword'):
                if obs.answers[i] == 'yes':
                    info += 'The keyword' + obs.questions[i][16:-1]  + '. '
                else:
                    info += 'The keyword does not' + obs.questions[i][16:-1]  + '. '
            elif obs.questions[i].startswith('May the keyword'):
                if obs.answers[i] == 'yes':
                    info += 'The keyword may' + obs.questions[i][16:-1]  + '. '
                else:
                    info += 'The keyword may not' + obs.questions[i][16:-1]  + '. '
            else:
                info += ' Answer for question "' + obs.questions[i] + '" is ' +  obs.answers[i]  + '. '
        for i in range(len(guesses)):
            info += f"It is not '{guesses[i]}'. "
            
        if q_list_counter < len(questions_list):
            response = questions_list[q_list_counter]
            q_list_counter += 1
        else:
            response = ask()
    elif obs.turnType == "guess":
        info = 'The keyword is a thing. '
        answers_length = len(obs.answers)
        for i in range(1, answers_length):
            if obs.questions[i].startswith('Is the keyword'):
                if obs.answers[i] == 'yes':
                    info += 'The keyword is' + obs.questions[i][14:-1]  + '. '
                else:
                    info += 'The keyword is not' + obs.questions[i][14:-1] + '. '
            elif obs.questions[i].startswith('Does the keyword'):
                if obs.answers[i] == 'yes':
                    info += 'The keyword' + obs.questions[i][16:-1]  + '. '
                else:
                    info += 'The keyword does not' + obs.questions[i][16:-1]  + '. '
            elif obs.questions[i].startswith('May the keyword'):
                if obs.answers[i] == 'yes':
                    info += 'The keyword may' + obs.questions[i][16:-1]  + '. '
                else:
                    info += 'The keyword may not' + obs.questions[i][16:-1]  + '. '
            else:
                info += ' Answer for question "' + obs.questions[i] + '" is ' +  obs.answers[i]  + '. '
     
        
        response = guess_default(obs.answers)
        guesses.append(response)
                
    else: #obs.turnType == "answer"
        HANDSHAKE = 'Is it Agent Alpha?'
        if obs.keyword.lower() in obs.questions[-1].lower():
            response = "yes"
        else:
            if HANDSHAKE == obs.questions[-1]:
                response = 'yes'
            else:
                testword = parse_alpha_question(obs.questions[-1])
                if testword is not None:
                    response = 'yes' if obs.keyword.lower() < testword else 'no'
                else:
                    response = get_yes_no(obs.questions[-1], obs.keyword)
            
    # DISPLAY ROLE
    if VERBOSE: 
        if obs.turnType == "answer": 
            print(f"Team 2 - Answerer - ### Agent LLAMA 8B ###")
        else:
            print(f"\nTeam 2 - Questioner - ### Agent LLAMA 8B ###")
        print(f"OUTPUT = '{response}'")

    return response


import random
import re

VERSION = 9
import pickle
with open(KAGGLE_AGENT_PATH + 'naive_216.pkl', 'rb') as f:
    wordlist = pickle.load(f)
for i in range(len(wordlist)):
    wordlist[i] = wordlist[i].lower()
# The list of all expected keywords in the game. It has to be updated before submitting your agent for the evaluation
# phase. Apply your own judgement. The longer the list the better coverage, but slower convergence. The coverage is
# arguably more important than convergence, so having many extra keywords is better than having missing words.
allwords = wordlist

last_alpha_yes = ''
last_alpha_no = ''


def answered_yes(obs, question):
    """ Check whether the question has been asked and the answer was 'yes'. """
    try:
        ix = obs.questions.index(question)
        return obs.answers[ix] == 'yes'
    except (ValueError, IndexError):
        return False


class AgentAlpha:
    global allwords
    # Keep it fixed, for compatibility
    HANDSHAKE = 'Is it Agent Alpha?'

    # This is our search space, it gets modified, so it needs a
    # reinit when testing with a series of games in a row
    keywords = sorted(allwords)

    @staticmethod
    def parse_alpha_question(question):
        # Regex to respond to alpha questions, includes all the 4 variants from the previous version:
        match = re.search(r'keyword.*(?:come before|precede) \"([^\"]+)\" .+ order\?$', question)
        if match:
            # Alpha question matched
            testword = match.group(1)
            return testword
        else:
            return None

    @staticmethod
    def play(obs, cfg):
        global last_alpha_yes, last_alpha_no, guesses, firstword, ALPHA_OPPONENT, pop_words
        if obs.turnType == 'ask':

            if len(obs.questions) == 0:
                return AgentAlpha.HANDSHAKE

            elif answered_yes(obs, AgentAlpha.HANDSHAKE) and AgentAlpha.keywords:
                ALPHA_OPPONENT = 'yes'
                testword = AgentAlpha.keywords[len(AgentAlpha.keywords) // 2]
                response = f'Does the keyword (in lowercase) precede "{testword}" in alphabetical order?'

            elif not AgentAlpha.keywords:
                response = 'No more alpha'
            else:
                response = None

        elif obs.turnType == 'answer':
            if AgentAlpha.HANDSHAKE == obs.questions[-1]:
                response = 'yes'
                #response = 'no' #declined

            else:
                testword = AgentAlpha.parse_alpha_question(obs.questions[-1])
                if testword is not None:
                    response = 'yes' if obs.keyword.lower() < testword else 'no'
                else:
                    # Not an alpha question, proceed with another approach
                    response = None

        elif obs.turnType == 'guess':

            # If the last question was a handshake, give a special guess
            if obs.questions[-1] == AgentAlpha.HANDSHAKE:
                # That's the handshake round...
                if obs.answers[-1] == 'yes':
                    ALPHA_OPPONENT = 'yes'
                    response = "plastic windows"
                else:
                    ALPHA_OPPONENT = 'no'
                    response = f'eye'

            else:
                # Bisect the space based on the last question and answer
                testword = AgentAlpha.parse_alpha_question(obs.questions[-1])
                if testword:
                    if obs.answers[-1] == 'yes':
                        last_alpha_yes = testword
                        AgentAlpha.keywords = [k for k in AgentAlpha.keywords if k < testword]
                    else:
                        last_alpha_no = testword
                        AgentAlpha.keywords = [k for k in AgentAlpha.keywords if not k < testword]
                firstword = last_alpha_no
                print('borders ', last_alpha_no, last_alpha_yes)
                if AgentAlpha.keywords:
                    # We pop from the front, but it doesn't matter much from which side we pop
                    if len(AgentAlpha.keywords) == 1 and (last_alpha_no == AgentAlpha.keywords[0] or last_alpha_yes == AgentAlpha.keywords[0]):
                        response = AgentAlpha.keywords.pop(0)
                    elif pop_words == False:
                        response = AgentAlpha.keywords[0]
                    else:
                        response = AgentAlpha.keywords.pop(0)
                else:
                    # Keywords exhausted, use something else
                    response = None
            guesses.append(response)
        else:
            assert False, f'Unexpected turnType: {obs.turnType}'

        return response
variants = []
def agent_dumb(obs, cfg):
    global variants
    # GENERATE RESPONSE
    if obs.turnType == "ask":
        response = f'Is it {variants[0]}?'
    elif obs.turnType == "guess":
        response = variants.pop(0)
    else: #obs.turnType == "answer"
        response = 'yes'
    return response

def agent_fn(obs, cfg):
    global variants, pop_words, guesses
    global ALPHA_OPPONENT, firstword, allwords, saved_word
    response = None
    if ALPHA_OPPONENT == '' or ALPHA_OPPONENT == 'yes':
        response = AgentAlpha.play(obs, cfg)
    if (response == 'No more alpha') or (ALPHA_OPPONENT == 'yes' and response is None) and not pop_words:
        #ALPHA_OPPONENT = 'dumb'
        print('NO MORE ALPHA')
        saved_word = firstword
        variants = search_pair()
        variants = [firstword + ' ' + variants[i].lower() for i in range(len(variants))]
        AgentAlpha.keywords = sorted(variants)
        pop_words = True
        response = AgentAlpha.play(obs, cfg)
    if pop_words and response is None:
        response = 'no'
    if response is None:
        response = agent_beta(obs, cfg)
    return response
```

```python

import json
import pandas as pd
#from tmp.submission.main import agent
from keywords_local import KEYWORDS_JSON

class Observation:
    def __init__(self):
        self.step = 0
        self.role = "guesser"
        self.turnType = "ask"
        self.keyword = "Japan"
        self.category = "country"
        self.questions = []
        self.answers = []
        self.guesses = []
        
def create_keyword_df(KEYWORDS_JSON):
    json_data = json.loads(KEYWORDS_JSON)

    keyword_list = []
    category_list = []
    alts_list = []

    for i in range(len(json_data)):
        for j in range(len(json_data[i]['words'])):
            keyword = json_data[i]['words'][j]['keyword']
            keyword_list.append(keyword)
            category_list.append(json_data[i]['category'])
            alts_list.append(json_data[i]['words'][j]['alts'])

    data_pd = pd.DataFrame(columns=['keyword', 'category', 'alts'])
    data_pd['keyword'] = keyword_list
    data_pd['category'] = category_list
    data_pd['alts'] = alts_list
    
    return data_pd
    
keywords_df = create_keyword_df(KEYWORDS_JSON)
```
